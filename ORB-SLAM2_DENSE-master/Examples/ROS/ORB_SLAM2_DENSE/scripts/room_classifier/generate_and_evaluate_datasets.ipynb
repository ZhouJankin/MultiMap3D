```python
##
# This module will create multiple data sets using room_labels_and_features_generator.py and then
# train SVCs on those data sets using room_classifier.py and collect the accuracies.
##

from room_classifier import RoomClassifier
from room_labels_and_features_generator import RoomClassifierTrainingDataGenerator
from ModelType import ModelType

import matplotlib.pyplot as plt
import numpy as np

def testGeneratedDataSet(model_type):
    # how many volumes do we want? We will increase each volume by 100. So if we want datasets of 
    # size [100, 200, 300] then this should be 3.
    number_of_dataset_volumes = 16

    # How many times we will re-generate data set with each volume? So if this was 5, 
    # and number_of_dataset_volumes = 3 then we would generate 5 data sets for each volume
    # of sizes [100, 200, 300] and we would get 15 tests altogether.
    number_of_tests_per_dataset_volume = 100

    # We'll create an empty Numpy array and put the our SVC accuracies into there.
    dataset_test_results = np.zeros(shape=(number_of_dataset_volumes, number_of_tests_per_dataset_volume))
    print(dataset_test_results.shape)

    for volume in range(number_of_dataset_volumes):
        accuracies = []
        for test in range(number_of_tests_per_dataset_volume):
            gen = RoomClassifierTrainingDataGenerator(model_type, (volume + 1) * 100)
            rc = RoomClassifier(True, model_type)
            accuracies.append(rc.getAccuracy())
        dataset_test_results[volume] = accuracies

    #print(dataset_test_results)
    fig, ax = plt.subplots()
    ax.axes.get_xaxis().set_visible(False)
    #ax7.set_title("Variance while " + operation + " " + gesture + " gesture")
    ax.set_title(model_type.name)
    ax.boxplot(dataset_test_results)
    plt.show()
    
testGeneratedDataSet(ModelType.FEATURES_12)
```
